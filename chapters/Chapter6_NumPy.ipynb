## 6.1 Why NumPy? A Speed Comparison

Before we dive into syntax, let's see why NumPy matters. We'll calculate the mean GPA two ways: with a Python list and loop, and with NumPy.

````python
# First, import NumPy (conventional alias: np)
import numpy as np

# Create sample GPA data (similar to our dataset)
gpa_list = [3.41, 3.55, 3.60, 3.44, 3.90, 4.0, 4.0, 4.0, 3.97, 2.41]

print("Sample GPAs:", gpa_list)
````

````python
# Method 1: Pure Python
def mean_python(data):
    total = 0
    for value in data:
        total += value
    return total / len(data)

print(f"Python mean: {mean_python(gpa_list):.4f}")
````

````python
# Method 2: NumPy
gpa_array = np.array(gpa_list)

print(f"NumPy mean: {np.mean(gpa_array):.4f}")
````

````python
# Speed comparison with larger data
import time

# Create 1 million random GPAs
large_list = [3.0 + (i % 100) / 100 for i in range(1_000_000)]
large_array = np.array(large_list)

# Time Python
start = time.time()
python_result = mean_python(large_list)
python_time = time.time() - start

# Time NumPy
start = time.time()
numpy_result = np.mean(large_array)
numpy_time = time.time() - start

print(f"Python: {python_time:.4f} seconds")
print(f"NumPy:  {numpy_time:.4f} seconds")
print(f"NumPy is {python_time / numpy_time:.0f}x faster")
````

NumPy achieves this speed because:
1. **Contiguous memory:** Arrays store values in adjacent memory locations
2. **Compiled code:** Operations are written in C, not interpreted Python
3. **Vectorization:** Operations on entire arrays at once, no Python loop overhead

For data analysis with thousands or millions of records, this difference is the difference between waiting seconds vs. minutes (or minutes vs. hours).

## 6.3 Indexing and Slicing

Array indexing works like list indexing, which you already know. The difference is that NumPy extends this to multiple dimensions.

````python
# 1D indexing (same as lists)
gpas = np.array([3.41, 3.55, 3.60, 3.44, 3.90, 4.0, 2.41, 3.11])

print(f"First GPA: {gpas[0]}")
print(f"Last GPA: {gpas[-1]}")
print(f"First three: {gpas[:3]}")
print(f"Every other: {gpas[::2]}")
````

````python
# 2D indexing
# Rows: students, Columns: [Credits_Attempted, Credits_Earned, GPA]
data = np.array([
    [105, 99, 3.41],   # Student 0
    [73, 62, 3.55],    # Student 1
    [89, 81, 3.60],    # Student 2
    [102, 98, 3.44],   # Student 3
    [52, 44, 3.90]     # Student 4
])

print("Full array:")
print(data)
````

````python
# Access single element: data[row, column]
print(f"Student 0's GPA: {data[0, 2]}")
print(f"Student 2's Credits Earned: {data[2, 1]}")
````

````python
# Access entire row
print(f"Student 0's data: {data[0]}")
print(f"Student 0's data (explicit): {data[0, :]}")
````

````python
# Access entire column
print(f"All GPAs: {data[:, 2]}")
print(f"All Credits Attempted: {data[:, 0]}")
````

````python
# Slice rows and columns
print("First 3 students, columns 1-2:")
print(data[:3, 1:3])
````

### Key Difference: Views vs. Copies

**Important:** NumPy slices create *views*, not copies. Modifying a view modifies the original array!

````python
# Demonstrate view behavior
original = np.array([1, 2, 3, 4, 5])
slice_view = original[:3]  # This is a VIEW, not a copy

print(f"Original: {original}")
print(f"Slice: {slice_view}")

# Modify the slice
slice_view[0] = 999

print(f"\nAfter modifying slice:")
print(f"Original: {original}")  # Original is also changed!
print(f"Slice: {slice_view}")
````

````python
# To make an independent copy, use .copy()
original = np.array([1, 2, 3, 4, 5])
slice_copy = original[:3].copy()  # This is an independent COPY

slice_copy[0] = 999

print(f"Original: {original}")   # Unchanged
print(f"Copy: {slice_copy}")
````

## 6.4 Element-wise Operations (Vectorization)

This is where NumPy shines. Operations are applied to every element simultaneously, no loops required.

````python
# Arithmetic with scalars
gpas = np.array([3.41, 3.55, 3.60, 3.44, 3.90])

# Add 0.1 bonus to all GPAs
boosted = gpas + 0.1
print(f"Original: {gpas}")
print(f"Boosted:  {boosted}")
````

````python
# Scale GPAs to percentage (multiply by 25)
gpa_percent = gpas * 25
print(f"GPA as percentage: {gpa_percent}")
````

````python
# Arithmetic between arrays (element by element)
credits_attempted = np.array([105, 73, 89, 102, 52])
credits_earned = np.array([99, 62, 81, 98, 44])

# Calculate completion rate for each student
completion_rate = credits_earned / credits_attempted
print(f"Completion rates: {completion_rate}")
````

````python
# Calculate credits lost
credits_lost = credits_attempted - credits_earned
print(f"Credits lost: {credits_lost}")
````

````python
# Compare to doing this with Python lists (DON'T do this in real code)
# This is just to show why vectorization matters

attempted_list = [105, 73, 89, 102, 52]
earned_list = [99, 62, 81, 98, 44]

# Python requires a loop or list comprehension
completion_python = [e / a for a, e in zip(attempted_list, earned_list)]
print(f"Python way: {completion_python}")

# NumPy: one operation, no loop
completion_numpy = credits_earned / credits_attempted
print(f"NumPy way:  {completion_numpy}")
````

## 6.5 Boolean Masking and Filtering

Comparison operators create boolean arrays, which can filter data. This is one of NumPy's most powerful features.

````python
gpas = np.array([3.41, 3.55, 3.60, 3.44, 3.90, 4.0, 2.41, 1.85])

# Comparison creates boolean array
high_gpa = gpas >= 3.5
print(f"GPAs: {gpas}")
print(f"GPA >= 3.5: {high_gpa}")
````

````python
# Use boolean array to filter (boolean masking)
honors_gpas = gpas[high_gpa]
print(f"Honors GPAs: {honors_gpas}")

# Or in one line:
print(f"Honors GPAs: {gpas[gpas >= 3.5]}")
````

````python
# Students on probation (GPA < 2.0)
probation = gpas[gpas < 2.0]
print(f"Probation GPAs: {probation}")
````

````python
# Combine conditions with & (and), | (or), ~ (not)
# NOTE: Must use parentheses around each condition!

# Students with GPA between 2.0 and 3.5 (inclusive)
middle_tier = gpas[(gpas >= 2.0) & (gpas <= 3.5)]
print(f"Middle tier (2.0-3.5): {middle_tier}")

# Students with very low OR very high GPA
extreme = gpas[(gpas < 2.0) | (gpas >= 3.9)]
print(f"Extreme GPAs: {extreme}")
````

````python
# Count how many meet a condition
print(f"Number of students >= 3.5: {np.sum(gpas >= 3.5)}")
print(f"Number on probation: {np.sum(gpas < 2.0)}")
````

````python
# np.where: get indices where condition is True
high_indices = np.where(gpas >= 3.5)
print(f"Indices of high GPAs: {high_indices[0]}")

# np.where with replacement values: condition ? true_value : false_value
status = np.where(gpas >= 2.0, "Good Standing", "Probation")
print(f"Status: {status}")
````

## 6.8 Broadcasting

Broadcasting is how NumPy handles operations between arrays of different shapes. It's powerful but can be confusing at first.

````python
# We've already seen broadcasting with scalars
gpas = np.array([3.41, 3.55, 3.60])
adjusted = gpas + 0.1  # 0.1 is "broadcast" to match gpas shape
print(adjusted)
````

````python
# Broadcasting with arrays
# Each row is a student: [Credits_Attempted, Credits_Earned]
credits = np.array([
    [105, 99],
    [73, 62],
    [89, 81]
])

# Multiply each column by a different weight
# First column (attempted) by 1.0, second (earned) by 1.2
weights = np.array([1.0, 1.2])

weighted = credits * weights
print("Original:")
print(credits)
print("\nWeighted:")
print(weighted)
````

How did that work? NumPy "stretched" the 1D weights array `[1.0, 1.2]` to match the 2D credits array:

```
credits (3x2):       weights (1x2 broadcast):      result:
[105, 99]      *     [1.0, 1.2]              =     [105.0, 118.8]
[73, 62]       *     [1.0, 1.2]              =     [73.0, 74.4]
[89, 81]       *     [1.0, 1.2]              =     [89.0, 97.2]
```

````python
# Broadcasting rule: dimensions must be equal OR one of them is 1
# Working backward from the rightmost dimension

a = np.ones((3, 4))     # Shape: (3, 4)
b = np.ones((4,))       # Shape: (4,) - broadcasts to (3, 4)
c = np.ones((3, 1))     # Shape: (3, 1) - broadcasts to (3, 4)

print(f"a + b shape: {(a + b).shape}")  # Works
print(f"a + c shape: {(a + c).shape}")  # Works
````

````python
# What doesn't work:
d = np.ones((3,))  # Shape: (3,)

try:
    result = a + d  # (3, 4) + (3,) - last dims don't match!
except ValueError as e:
    print(f"Error: {e}")
````

### Practical Example: Standardizing Data

````python
# Standardize: subtract mean, divide by std (z-score normalization)
# This centers data at 0 with std of 1

gpas = np.array([3.41, 3.55, 3.60, 3.44, 3.90, 4.0, 2.41, 1.85])

mean_gpa = np.mean(gpas)
std_gpa = np.std(gpas)

# Broadcasting in action: subtract scalar, divide by scalar
standardized = (gpas - mean_gpa) / std_gpa

print(f"Original GPAs: {gpas}")
print(f"Standardized: {np.round(standardized, 2)}")
print(f"New mean: {np.mean(standardized):.6f}")  # Should be ~0
print(f"New std: {np.std(standardized):.6f}")    # Should be ~1
````

## 6.9 Axis Parameter

For 2D arrays, you often want to aggregate along rows OR columns. The `axis` parameter controls this.

- `axis=0`: Operate along rows (result has one value per column)
- `axis=1`: Operate along columns (result has one value per row)

**Memory trick:** The axis number is the dimension that "collapses."

````python
# Student data: rows = students, columns = [Credits_Attempted, Credits_Earned, GPA]
data = np.array([
    [105, 99, 3.41],   # Student 0
    [73, 62, 3.55],    # Student 1
    [89, 81, 3.60],    # Student 2
    [102, 98, 3.44],   # Student 3
])

print("Data (4 students, 3 features):")
print(data)
print(f"Shape: {data.shape}")
````

````python
# No axis: aggregate all elements
print(f"Mean of all values: {np.mean(data):.2f}")
````

````python
# axis=0: collapse rows -> one value per column
# "Mean down the rows" = column means
col_means = np.mean(data, axis=0)
print(f"Column means (axis=0): {col_means}")
print("Interpretation: [avg credits attempted, avg credits earned, avg GPA]")
````

````python
# axis=1: collapse columns -> one value per row
# "Mean across the columns" = row means (not very meaningful here)
row_means = np.mean(data, axis=1)
print(f"Row means (axis=1): {row_means}")
print("Interpretation: [student0 avg, student1 avg, ...]")
````

````python
# More meaningful example: sum credits by student
# Sum columns 0 and 1 (credits attempted + credits earned) for each row
credits_data = data[:, :2]  # First two columns only
total_credits_per_student = np.sum(credits_data, axis=1)
print(f"Total credits per student: {total_credits_per_student}")
````

````python
# Min and max by column
print(f"Min by column: {np.min(data, axis=0)}")
print(f"Max by column: {np.max(data, axis=0)}")
````

## 6.10 Reshaping Arrays

Sometimes you need to change the shape of an array without changing its data.

````python
# Create 1D array
flat = np.arange(12)
print(f"Flat: {flat}")
print(f"Shape: {flat.shape}")
````

````python
# Reshape to 3x4
matrix = flat.reshape(3, 4)
print("Reshaped to 3x4:")
print(matrix)
````

````python
# Reshape to 4x3
matrix2 = flat.reshape(4, 3)
print("Reshaped to 4x3:")
print(matrix2)
````

````python
# Use -1 to infer dimension
# "I want 4 rows, figure out the columns"
auto = flat.reshape(4, -1)
print(f"Shape with -1: {auto.shape}")
````

````python
# Flatten back to 1D
flat_again = matrix.flatten()
print(f"Flattened: {flat_again}")

# Or use ravel() - returns view when possible (more efficient)
flat_view = matrix.ravel()
print(f"Raveled: {flat_view}")
````

````python
# Transpose: swap rows and columns
print("Original (3x4):")
print(matrix)
print(f"\nTransposed (4x3):")
print(matrix.T)
````

## 6.11 Combining Arrays

````python
# Vertical stacking (add rows)
batch1 = np.array([[1, 2], [3, 4]])
batch2 = np.array([[5, 6], [7, 8]])

combined = np.vstack([batch1, batch2])
print("Vertical stack:")
print(combined)
````

````python
# Horizontal stacking (add columns)
combined_h = np.hstack([batch1, batch2])
print("Horizontal stack:")
print(combined_h)
````

````python
# Practical: add a new column to our student data
data = np.array([
    [105, 99],
    [73, 62],
    [89, 81],
])

# Calculate completion rate as new column
completion = (data[:, 1] / data[:, 0]).reshape(-1, 1)  # Make it a column

# Add to data
data_with_completion = np.hstack([data, completion])
print("Data with completion rate:")
print(data_with_completion)
````
